## 3.5 Explore the Relationship between the columns and try to answer the following questions:

1. Is there any significant difference between men/women's salary?

2. Is there any significant difference between men/women's number of the purchase in the last three years?

```python
from scipy.stats import ttest_ind
```

```python
#Question 1
Men_Ann_Salary = data[data.Gender == 'M']['Cust_Ann_Income']
Women_Ann_Salary = data[data.Gender == 'F']['Cust_Ann_Income']

p_value_Ann_Salary = ttest_ind(Men_Ann_Salary, Women_Ann_Salary, equal_var = False)[1]

if p_value_Ann_Salary < 0.05:
    print("""
          P_value = {}
           We can REJECT the null hypothesis and conclude the alternative hypothesis:
          Men and Women's Annual Salary are different""".format(round(p_value_Ann_Salary, 4)))
else:
    print("""
          P_value = {}
          We CANT REJECT the null hypothesis, as there are not enough evidence to support the alternative hypothesis for the alternative hypothesis that there is a relationship between Gender and the Annual Salary""".format(round(p_value_Ann_Salary,4)))
```

```python
#Question 2
Men_Pur_3 = data[data.Gender == 'M']['Pur_3_years']
Women_Pur_3 = data[data.Gender == 'F']['Pur_3_years']

p_value_Pur_3 = ttest_ind(Men_Pur_3, Women_Pur_3, equal_var = False)[1]

if p_value_Pur_3 < 0.05:
    print("""
          P_value = {}
          We can REJECT the null hypothesis and conclude the alternative hypothesis:
          Men and Women's amount of purchases in the last 3 years are different""".format(round(p_value_Pur_3, 4)))
else:
    print("""
          P_value = {}
          We CANT REJECT the null hypothesis, as there are not enough evidence to support the alternative hypothesis for the alternative hypothesis that there is a relationship between Gender and the amount of purchases in the last 3 years""".format(round(p_value_Pur_3,4)))
```

```python
def best_model(X_train, X_test, Y_train, Y_test, model, params, CV = 5, **kwargs):
    param_grid = params
    grid = GridSearchCV(model,param_grid,cv = CV,
                        scoring = 'roc_auc', return_train_score = True)
    grid.fit(X_train, Y_train)
    best_par = grid.best_params_
    best_model = grid.best_estimator_.fit(X_train, Y_train)
    Y_pred = best_model.predict(X_test)
    f1 = f1_score(Y_test, Y_pred)
    best_sc = grid.best_score_
    conf_matrix = confusion_matrix(Y_test, Y_pred)
    class_report = classification_report(Y_test, Y_pred)
    return best_par, best_sc, f1, conf_matrix, class_report

```

```python
def assigner(Y_test, Y_pred):
    tp_logic = (Y_test == 1) & (Y_pred == 1)
    tn_logic = (Y_test == 0) & (Y_pred == 0)
    fp_logic = (Y_test == 0) & (Y_pred == 1)
    fn_logic = (Y_test == 1) & (Y_pred == 0)
    return np.sum(tp_logic), np.sum(tn_logic), np.sum(fp_logic), np.sum(fn_logic)

knn_tp_0, knn_tn_0, knn_fp_0, knn_fn_0 = assigner(Y_test, knn_pred_0)
knn_tp_1, knn_tn_1, knn_fp_1, knn_fn_1 = assigner(Y_test, knn_pred_1)
knn_tp_2, knn_tn_2, knn_fp_2, knn_fn_2 = assigner(Y_test, knn_pred_2)

dt_tp_0, dt_tn_0, dt_fp_0, dt_fn_0 = assigner(Y_test, dt_pred_0)
dt_tp_1, dt_tn_1, dt_fp_1, dt_fn_1 = assigner(Y_test, dt_pred_1)
dt_tp_2, dt_tn_2, dt_fp_2, dt_fn_2 = assigner(Y_test, dt_pred_2)

lr_tp_0, lr_tn_0, lr_fp_0, lr_fn_0 = assigner(Y_test, lr_pred_0)
lr_tp_1, lr_tn_1, lr_fp_1, lr_fn_1 = assigner(Y_test, lr_pred_1)
lr_tp_2, lr_tn_2, lr_fp_2, lr_fn_2 = assigner(Y_test, lr_pred_2)

svm_tp_0, svm_tn_0, svm_fp_0, svm_fn_0 = assigner(Y_test, svm_pred_0)
svm_tp_1, svm_tn_1, svm_fp_1, svm_fn_1 = assigner(Y_test, svm_pred_1)
#svm_tp_2, svm_tn_2, svm_fp_2, svm_fn_2 = assigner(Y_test, svm_pred_2)

def calculations(avg_pur, tp, fn, fp):
    results_cost = 5 * (tp + fp)
    results_lost =  avg_pur * fn
    results_gain = avg_pur * tp
    results_value = results_gain - results_lost - results_cost
    results_revenue = results_gain - results_cost
    return results_value, results_revenue

knn_value_0, knn_rev_0 = calculations(avg_pur, knn_tp_0, knn_fn_0, knn_fp_0)
knn_value_1, knn_rev_1 = calculations(avg_pur, knn_tp_1, knn_fn_1, knn_fp_1)
knn_value_2, knn_rev_2 = calculations(avg_pur, knn_tp_2, knn_fn_2, knn_fp_2)

dt_value_0, dt_rev_0 = calculations(avg_pur, dt_tp_0, dt_fn_0, dt_fp_0)
dt_value_1, dt_rev_1 = calculations(avg_pur, dt_tp_1, dt_fn_1, dt_fp_1)
dt_value_2, dt_rev_2 = calculations(avg_pur, dt_tp_2, dt_fn_2, dt_fp_2)

lr_value_0, lr_rev_0 = calculations(avg_pur, lr_tp_0, lr_fn_0, lr_fp_0)
lr_value_1, lr_rev_1 = calculations(avg_pur, lr_tp_1, lr_fn_1, lr_fp_1)
lr_value_2, lr_rev_2 = calculations(avg_pur, lr_tp_2, lr_fn_2, lr_fp_2)

svm_value_0, svm_rev_0 = calculations(avg_pur, svm_tp_0, svm_fn_0, svm_fp_0)
svm_value_1, svm_rev_1 = calculations(avg_pur, svm_tp_1, svm_fn_1, svm_fp_1)
#svm_value_2, svm_rev_2 = calculations(avg_pur, svm_tp_2, svm_fn_2, svm_fp_2)

values = pd.Series([knn_value_0, knn_value_1, knn_value_2, dt_value_0, dt_value_1, dt_value_2,
                    lr_value_0, lr_value_1, lr_value_2, svm_value_0, svm_value_1])
revs = pd.Series([knn_rev_0, knn_rev_1, knn_rev_2, dt_rev_0, dt_rev_1, dt_rev_2, 
                  lr_rev_0, lr_rev_1, lr_rev_2, svm_rev_0, svm_rev_1])
```
